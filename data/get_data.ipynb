```python
import yfinance as yf
import os

# 1. Define the currency and period
TICKER = "BTC-USD"  
START_DATE = "2020-01-01"
END_DATE = "2024-12-01"

# 2. Download the data
print(f"Downloading data for {TICKER}...")
df = yf.download(TICKER, start=START_DATE, end=END_DATE)

# 3. Reset the index to have 'Date' as a normal column
df.reset_index(inplace=True)

# 4. Check the folder structure
if not os.path.exists('raw'):
    os.makedirs('raw')

# 5. Save the raw CSV file
csv_path = 'raw/bitcoin_raw.csv'
df.to_csv(csv_path, index=False)

print(f"Success! The data has been saved in '{csv_path}'")
print(f"We have {df.shape[0]} rows and {df.shape[1]} columns.")
print(df.head())
```
